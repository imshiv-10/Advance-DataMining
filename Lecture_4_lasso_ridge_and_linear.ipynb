```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Lasso, Ridge

# Generate dataset
X, y = make_regression(n_samples=50, n_features=1, noise=20)

# Linear regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

# Lasso regression
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
y_pred_lasso = lasso.predict(X)

# Ridge regression
ridge = Ridge(alpha=0.1)
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

# Plot results
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_lin, color='red', label='Linear Regression')
plt.plot(X, y_pred_lasso, color='green', label='Lasso Regression')
plt.plot(X, y_pred_ridge, color='purple', label='Ridge Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Lasso, Ridge

# Generate dataset
X, y = make_regression(n_samples=50, n_features=10, noise=20)

# Linear regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

# Lasso regression
lasso = Lasso(alpha=10, max_iter=10000)
lasso.fit(X, y)
y_pred_lasso = lasso.predict(X)

# Ridge regression
ridge = Ridge(alpha=10)
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

# # Plot results
# plt.scatter(X[:,0], y, color='blue')
# plt.plot(X[:,0], y_pred_lin, color='red', label='Linear Regression')
# plt.plot(X[:,0], y_pred_lasso, color='green', label='Lasso Regression')
# plt.plot(X[:,0], y_pred_ridge, color='purple', label='Ridge Regression')
# plt.legend()
# plt.show()

# Print the selected features by Lasso
print("Selected features by Lasso:", np.where(lasso.coef_ != 0)[0])

# Print the distribution of feature weights of each model
print("Linear Regression feature weights:", lin_reg.coef_)
print("Lasso Regression feature weights:", lasso.coef_)
print("Ridge Regression feature weights:", ridge.coef_)

# Plot histogram of feature weights
plt.hist(lin_reg.coef_, bins = 20, color = 'blue', label='Linear Regression')
plt.legend()
plt.show()
plt.hist(lasso.coef_, bins = 20, color = 'green', label='Lasso Regression')
plt.legend()
plt.show()
plt.hist(ridge.coef_, bins = 20, color = 'purple', label='Ridge Regression')
plt.legend()
plt.show()
```
